```python
import pandas as pd
import ast
from sklearn.feature_extraction.text import CountVectorizer

# ==============================
# 1. Load CSV files
# ==============================
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")

# Merge on 'title'
movies = movies.merge(credits, on='title')

# Keep required columns
movies = movies[['id','title','overview','genres','keywords','cast','crew']]

# ==============================
# 2. Handle missing values
# ==============================
movies.dropna(inplace=True)

# ==============================
# 3. Convert genres & keywords to lists
# ==============================
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L

movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)

# ==============================
# 4. Convert top 3 cast members
# ==============================
def convert_cast(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter < 3:
            L.append(i['name'])
        counter += 1
    return L

movies['cast'] = movies['cast'].apply(convert_cast)

# ==============================
# 5. Extract director
# ==============================
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L

movies['crew'] = movies['crew'].apply(fetch_director)

# ==============================
# 6. Split overview into words
# ==============================
movies['overview'] = movies['overview'].apply(lambda x: x.split())

# ==============================
# 7. Create 'tags' column
# ==============================
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']

# Combine lists into a single string per movie
movies['tags'] = movies['tags'].apply(lambda x: " ".join(x).lower())

# ==============================
# 8. CountVectorizer
# ==============================
cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(movies['tags']).toarray()

# ==============================
# 9. Check results
# ==============================
feature_names = cv.get_feature_names_out()
print("Number of features:", len(feature_names))   # Should be close to 5000
print("Shape of vectors:", vectors.shape)          # (number_of_movies, 5000)
print("First 10 features:", feature_names[:10])
```

```python
import pickle
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load movies.pkl
movies = pickle.load(open('movies.pkl', 'rb'))

# --- Step 1: Create 'tags' column ---
# tags = overview + genres + keywords + cast + crew
# But in movies.pkl, usually 'overview','genres','keywords','cast','crew' already exist as strings/lists

# First combine into one string
def combine_features(row):
    return row['overview'] + " " + row['genres'] + " " + row['keywords'] + " " + row['cast'] + " " + row['crew']

movies['tags'] = movies.apply(combine_features, axis=1)

# Convert all to lowercase
movies['tags'] = movies['tags'].apply(lambda x: x.lower())

# --- Step 2: Vectorization ---
cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(movies['tags']).toarray()

# --- Step 3: Cosine Similarity ---
similarity = cosine_similarity(vectors)

# --- Step 4: Save similarity.pkl ---
pickle.dump(similarity, open('similarity.pkl', 'wb'))

print("✅ similarity.pkl created successfully!")
```
